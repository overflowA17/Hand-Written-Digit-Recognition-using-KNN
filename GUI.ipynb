```python
import tkinter as tk
import numpy as np
from PIL import ImageTk, Image, ImageDraw
import cv2
import joblib
```

```python
win = tk.Tk()#main window

model = joblib.load('KNN_handwritten_digits.sav')

fontButton = 'Arial 20 bold'
count = 0

def eventFunction(event):
    
    x = event.x
    y = event.y
    
    x1 = x - 20
    y1 = y - 20
    x2 = x + 20
    y2 = y + 20
    
    canvas.create_oval((x1, y1, x2, y2), fill = 'black')
    imgDraw.ellipse((x1, y1, x2, y2), fill = 'white')
    

def save():
    
    global count
    
    imgArray = np.array(img)
    imgArray = cv2.resize(imgArray, (8, 8))
    
    cv2.imwrite('Data/' + str(count) + '.jpg', imgArray)
    count = count + 1

    

def clear():
    
    global img, imgDraw
    
    canvas.delete('all')
    img = Image.new('RGB', (500, 500), (0, 0, 0))
    imgDraw = ImageDraw.Draw(img)
    

def predict():
    
    imgArray = np.array(img)
    
    imgArray = cv2.cvtColor(imgArray, cv2.COLOR_RGB2GRAY)
    
    imgArray = cv2.resize(imgArray, (8, 8))
    imgArray = np.reshape(imgArray, (1, 64))
    imgArray = (imgArray/255.0) * 15.0
    
    
    result = model.predict(imgArray)
    
    labelStatus.config(text = 'PREDICTED DIGIT: ' + str(result))
    
#canvas
canvas = tk.Canvas(win, width = 500, height = 500, bg = 'white')
canvas.grid(row = 0, column = 0, columnspan = 4)

canvas.bind('<B1-Motion>', eventFunction)

#buttons
button_save = tk.Button(win, text  = 'SAVE', bg = 'green', fg = 'white', font = fontButton, command = save)
button_save.grid(row = 1, column = 0)

button_predict = tk.Button(win, text  = 'PREDIICT', bg = 'blue', fg = 'white', font = fontButton, command = predict)
button_predict.grid(row = 1, column = 1)

button_clear = tk.Button(win, text  = 'CLEAR', bg = 'darkgoldenrod1', fg = 'white', font = fontButton, command = clear)
button_clear.grid(row = 1, column = 2)
                    
button_exit = tk.Button(win, text  = 'EXIT', bg = 'red', fg = 'white', font = fontButton, command = win.destroy)
button_exit.grid(row = 1, column = 3)

labelStatus = tk.Label(win, text = 'PREDICTED DIGIT: NONE', bg = 'black', fg = 'white', font = fontButton)
labelStatus.grid(row = 2, column = 0, columnspan = 4)


img = Image.new('RGB', (500, 500), (0, 0, 0))
imgDraw = ImageDraw.Draw(img)

win.mainloop()
```
